# Importing the Data Set


```
### RUN THIS FIRST TO IMPORT DATA ###

import pandas as pd
import numpy as np
import numpy.linalg as npl
import matplotlib.pyplot as plt

ts = pd.read_csv('/content/New_Cases_TimeSeries.csv')
data = ts.values

# n is number of days for which we have data on case numbers
# First four fields of each row in csv file are geographical metadata
n = np.size(data,1) - 4

# t will be the horizontal axis for many of the following time series plots
t = np.arange(n)
```

## Step one

Graphing the states actual values over the given period of time

```
# Graphing values #

states = list()
for st in data:
  if st[1] not in states:
    states.append(st[1])
  

print(states)
for ind in states:
  inc = np.where(data==ind)
  act = data[inc[0]]

  allVals = (inc[0])

  n = len(inc)
  total = []

  for each in allVals:
    print(data[each])
    total.insert(0, data[each][4])

  print("----------"+ind+"----------")
  plt.plot(total)
  plt.show()  


```

## Step Two
Fitting the data to a polynomial model

```
# Fitting the model

for st in states:

  inc = np.where(data==st)
  act = data[inc[0]]

  allVals = (inc[0])

  n = len(inc)
  total = []

  for each in allVals:
    #print(data[each][4])
    total.insert(0, data[each][4])

  total.pop(0)


  caseS = np.array(total, dtype='float')
  print(caseS)
  n = len(total)
  x = np.arange(n)
  X = np.column_stack((np.ones(n), x, np.power(x,2), np.power(x,3), np.power(x,4), np.power(x,5), np.power(x,6)))

  sol = npl.lstsq(X, caseS, rcond=None)
  beta = sol[0]
  SSE = sol[1]

  # Overlay trend line on data

  print("----------"+st+"----------")
  plt.scatter(np.arange(n), caseS)
  plt.plot(np.arange(n), X @ beta, 'r')
  print("beta =", beta)
  print("RMS = ", np.sqrt(SSE/n))
  plt.show()

```

## Step Three
Predicting the up coming days for all the states using an autoregressive model

*All fitted to the same model, not the most accurate

```
# Predicting #


for st in states:


  inc = np.where(data==st)
  act = data[inc[0]]

  allVals = (inc[0])

  n = len(inc)
  total = []

  for each in allVals:
    #print(data[each][4])
    total.insert(0, data[each][4])

  total.pop(0)

  temps = np.array(total, dtype='float')

  N = len(temps)
  print(N)

  #plt.scatter(np.arange(N), temps)

  M = 7

  x2 = range(M,N+6)

  # Observation vector y is just all observations except first M

  # once you have your initial predict matrix, intitial 
  y = temps[M:]

  X = temps[0:N-M]
  for i in range(M-1):
    X = np.column_stack((X, temps[i+1:N-M+i+1]))

  sol = npl.lstsq(X, y, rcond=None)

  pre = temps[-M:]
  print(pre)

  # The predict part
  for i in range(5):
    theta = sol[0]
    yhat = pre @ theta
    #print("yhat", yhat)
    pre = list(pre)
    pre.pop(0)
    pre.append(yhat)
  pre.pop(0)


  beta = sol[0]
  SSE = sol[1]


  l = list(X @ beta)
  tots = l + pre

  # dist = range(N,N+6)

  print("----------"+st+"----------")

  plt.scatter(np.arange(N), total)
  # plt.plot(np.arange(M,N), (X @ beta), color='green')
  # plt.plot(dist, pre, color='red')

  plt.plot(x2[:-5], tots[:-5], color = "green")
  plt.plot(x2[-6:], tots[-6:], color = "red")

  print("beta =", beta)
  print("RMS = ", np.sqrt(SSE/len(y)))


  # DO NOT MODIFY #
  plt.show()
  # DO NOT MODIFY #
  plt.show()


```
